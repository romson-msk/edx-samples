### Задача №1. Робинзон
#### *Условие: принять 1 мес = 30 дней (1 год = 12х30= 360 дней)*

```python
k = int(input("Всего дней на острове?"))

y = k // 360
m = (k - (y * 360)) // 30
w = (k - (y * 360) - (m * 30)) // 7
d = (k - (y * 360) - (m * 30) - (w * 7))

print("Дней всего (из расчета 1 мес = 30 дн):", k, "дней - это будет: ", y, "л.", m, "мес.", w, "нед.", d, "дн.")
print("Проверка =", y * 360 + m * 30 + w * 7 + d)
```

### Задача №2. Часы

```python
n = int(input("Сначала суток прошло минут?"))

if n <= 1440: # не более суток прошло
    clock_h = n // 60
    clock_m = n - clock_h * 60   
else:
    clock_h = (n - (n // 1440) * 1440) // 60
    clock_m = (n - (n // 1440) * 1440) - clock_h * 60

print("Время на часах сейчас:", clock_h, "ч.", clock_m, "мин.")
```
